```python
data = {'A':[2,5,np.nan,8,np.nan,9],
       'B':[np.nan,45,np.nan,89,63,np.nan], 
       'C':[np.nan,74,np.nan,np.nan,85,4], 
       'D':[10,20,30,40,50,60]}
data          #non
```

```python
df2 = pd.DataFrame(data)
df2
```

how to handle missing value 
     1. Remove the records 
     2. fill the records 
     

## removing the records

```python
df2.dropna()        #by default remove all those rows that contain missing value
```

```python
df2.dropna(axis='columns')
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
print("total missing value in your df :",df2.isnull().sum().sum())
```

```python
df2.isnull().sum()/df2.shape[0]*100
#percentage amount 
```

## filling the records

before filling 
   1.  check column is numerical or categorical
          if numerical then 
                              1. fill constant value (same value at every missing value)
                              2. mean 
                              3. median
          if categorical 
                         1. constant 
                         2. mode 


```python
df2
```

```python
# df2.fillna(500)  in full dataframe same value 

# for filling column wise
df2['A'].fillna(500)
df2['B'].fillna(600)
df2['C'].fillna(700)
```

```python
df2['A'].mean()
```

```python
df2.fillna(df2['A'].mean())
```

```python
df2.fillna(df2['A'].median())
```
